```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib
import numpy as np
import os
```

```python
def evaluate(X: np.ndarray, out_dir: str = 'outputs', seed: int = 42):
    os.makedirs(out_dir, exist_ok=True)

    scaler = StandardScaler()
    X_test_scaled = scaler.fit_transform(X)
    model = joblib.load('/content/xgb_regressor75k.joblib')
    y_pred = model.predict(X_test_scaled)
    df_old = pd.read_csv('/content/test.csv')
    df_new = pd.DataFrame({
    'sample_id': df_old['sample_id'],
    'y_pred': y_pred
    })
    return df_new
```

```python
X = np.load('/content/fused_x 75ktest.npy')
```

```python
output_df = evaluate(X)
```

```python
output_df.to_csv('test_out.csv',index=False)
```
