```python
# --- STEP 1: Mount Google Drive ---
from google.colab import drive
drive.mount('/content/drive')

```

```python
# --- STEP 2: Import libraries ---
import pandas as pd
import requests
import os
from tqdm import tqdm


```

```python
# --- STEP 3: Load CSV file ---
csv_path = "/content/test.csv"  # Update if needed
df = pd.read_csv(csv_path)

```

```python
# --- STEP 4: Create save directory inside Google Drive ---
save_dir = "/content/drive/MyDrive/images_test"
os.makedirs(save_dir, exist_ok=True)

```

```python
# --- STEP 5: Stream images directly to Drive ---
print("ðŸ“¥ Downloading images directly to Google Drive...")

for _, row in tqdm(df.iterrows(), total=len(df), desc="Downloading", ncols=100):
    img_url = str(row['image_link']).strip()
    sample_id = str(row['sample_id']).strip()
    save_path = os.path.join(save_dir, f"{sample_id}.jpg")

    # Skip already downloaded
    if os.path.exists(save_path):
        continue

    try:
        response = requests.get(img_url, stream=True, timeout=10)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
        else:
            print(f"âš ï¸ Failed: {img_url} ({response.status_code})")
    except Exception as e:
        print(f"âŒ Error downloading {img_url}: {e}")

print("âœ… All images saved directly to Google Drive:", save_dir)
```
